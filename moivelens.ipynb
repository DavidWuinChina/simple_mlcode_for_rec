```python
import pandas as pd  # 导入Pandas库，用于数据处理和分析
import numpy as np   # 导入Numpy库，用于数值计算
```

```python
dtype = {
    "userId": np.int32,  # 用户ID数据类型设置为32位整数
    "movieId": np.int32,  # 电影ID数据类型设置为32位整数
    "rating": np.float32  # 评分数据类型设置为32位浮点数
}
ratings = pd.read_csv(r'C:\python文件\推荐算法实例\ml-latest-small\ml-latest-small\ratings.csv', dtype=dtype, usecols=range(3))  
```

```python
ratings
```

```python
rating_matrix=ratings.pivot_table(index='userId',columns='movieId',values='rating')
```

```python
rating_matrix
```

```python
similarity_matrix=rating_matrix.T.corr()
similarity_matrix
```

```python
similar_user = similarity_matrix.iloc[1].drop(index=1).dropna()
similar_user
```

```python
similar_user=similar_user.where(similar_user>0).dropna()
similar_user
```

```python
# 获取 rating_matrix 中非缺失值的索引
rating_indices = rating_matrix[1].dropna().index

# 获取 similar_user 的索引
similar_indices = similar_user.index

# 找到两个索引的交集
ids = rating_indices.intersection(similar_indices)
ids
```

```python
final_similar_user = similar_user.loc[ids]
final_similar_user
```

```python
sum_up = 0  # 存储评分的加权和
sum_down = 0  # 存储权重的和
for sim_uid, similarity in final_similar_user.items():
    sim_user_rated_movies = rating_matrix.loc[sim_uid].dropna()
    sim_user_rating_for_item = sim_user_rated_movies[1]
    sum_up += similarity * sim_user_rating_for_item
    sum_down += similarity
predict_rating = sum_up / sum_down
print("用户1对物品1预测评分:",predict_rating)
```

```python
def predict(uid, iid, ratings_matrix, user_similar):
    '''
    预测给定用户对给定物品的评分值
    :param uid: 用户ID
    :param iid: 物品ID
    :param ratings_matrix: 用户-物品评分矩阵
    :param user_similar: 用户间相似度矩阵
    :return: 预测的评分值
    '''
    print("开始预测用户<%d>对电影<%d>的评分..." % (uid, iid))

    # 1. 找出uid用户的相似用户
    similar_users = user_similar[uid].drop([uid]).dropna()

    # 相关用户筛选规则：正相关的用户
    similar_users = similar_users.where(similar_users > 0).dropna()
    if similar_users.empty:
        raise Exception("用户<%d>没有相似的用户" % uid)

    # 2. 从uid用户的近邻用户中筛选出对iid物品有评分记录的近邻用户
    ids = set(ratings_matrix[iid].dropna().index) & set(similar_users.index)
    finally_similar_users = similar_users.loc[list(ids)]

    # 3. 结合uid用户与其邻近用户的相似度预测uid用户对iid物品的评分
    sum_up = 0  # 评分预测公式的分子部分的值
    sum_down = 0  # 评分预测公式的分母部分的值
    for sim_uid, similarity in finally_similar_users.items():  # 修改此行
        # 近邻用户的评分数据
        sim_user_rated_movies = ratings_matrix.loc[sim_uid].dropna()
        
        # 近邻用户对目标物品的评分
        sim_user_rating_for_item = sim_user_rated_movies[iid]
    
        # 计算分子的值
        sum_up += similarity * sim_user_rating_for_item
        # 计算分母的值
        sum_down += similarity
    
    # 计算预测的评分值并返回
    if sum_down == 0:
        return 0  # 防止除以零
    else:
        predict_rating = sum_up / sum_down
        print("预测出用户<%d>对电影<%d>的评分：%.2f" % (uid, iid, predict_rating))
        return round(predict_rating, 2)

# 调用函数
predict(1, 2, rating_matrix, similarity_matrix)
```

```python
def predict_all(uid, ratings_matrix, user_similar):
    '''
    预测全部评分
    :param uid: 用户ID
    :param ratings_matrix: 用户-物品评分矩阵
    :param user_similar: 用户间的相似度
    :return: 生成器，逐个返回预测评分
    '''
    # 准备要预测的物品的id列表
    item_ids = ratings_matrix.columns
    
    # 遍历预测
    for iid in item_ids:
        try:
            rating = predict(uid, iid, ratings_matrix, user_similar)
        except Exception as e:
            print(e)
        else:
            yield uid, iid, rating
def predict_user(x, ratings_matrix, user_similar):
    for i in predict_all(x, ratings_matrix, user_similar):
        print(i)
```

```python
# 调用函数
predict_user(2, rating_matrix, similarity_matrix)
```
